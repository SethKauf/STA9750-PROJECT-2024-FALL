```python
# import libraries
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib.ticker as ticker
import seaborn as sns
```

```python
# read in usage and financial data
df = pd.read_csv('data/usage_and_financials_2022.csv')
```

```python
dfs = df[['agency','mode','UPT','VRM']].groupby(['agency','mode']).max(['UPT','VRM']).reset_index()
```

```python
dfs['UPT'].describe()
```

```python
# Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(dfs['UPT'], bins=500, color='skyblue', edgecolor='black')

# Set x-axis to use regular formatting
plt.xscale('log')

plt.xlabel('UPT')
plt.ylabel('Frequency')
plt.title('Distribution of UPT')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

```python
# Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(dfs['VRM'], bins=500, color='skyblue', edgecolor='black')

# Set x-axis to use regular formatting
plt.xscale('log')

plt.xlabel('VRM')
plt.ylabel('Frequency')
plt.title('Distribution of VRM')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.show()
```

both heavily left-skewed with massive outliers
